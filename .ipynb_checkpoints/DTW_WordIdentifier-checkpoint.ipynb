## Implementasi Word Iden Pendekatan DTW Sederhana

```python
import time
import librosa
from dtw import dtw
import librosa.display
from scipy.spatial.distance import cdist
import matplotlib.pyplot as plt
import os
import numpy as np
import pandas as pd
```

```python
# load 4 audio from memory
audio1, scl1 = librosa.load('train/0_george_4.wav')
audio2, scl2 = librosa.load('train/0_george_5.wav')
audio3, scl3 = librosa.load('train/1_george_1.wav')
audio4, scl4 = librosa.load('train/1_george_4.wav')
```

```python
train_path = "train"

# Baca semua file
files = [file for file in os.listdir(train_path)]

# Buat matrix isi dengan 0 dan cari distance tiap audio
cost_matrix = np.ones((len(files), len(files)))

# Untuk label prediksi
y = np.ones(len(files))
durasi = []

# Iterasi untuk mencari cost antar tiap audio
for i in range(len(files)):
    audio1, scl1 = librosa.load(train_path + "/" + files[i])
    mfcc1 = librosa.feature.mfcc(audio1, scl1)
    durasi.append(librosa.get_duration(y=audio1, sr=scl1))
    for j in range (len(files)):
        audio2, scl2 = librosa.load(train_path + '/' + files[j])
        mfcc2 = librosa.feature.mfcc(audio2, scl2)
        _,_,cost = dtw(mfcc1.T, mfcc2.T, dist=lambda x, y: np.linalg.norm(x-y, ord=1))
        cost_matrix[(i,j)] = cost
    # Indicate audio of "zero"
    if (i<=10):
        y[i] = 0 
    # Indicate audio of "one"
    elif (10<i<=21):
        y[i] = 1
    # Indicate audio of "two"
    else:
        y[i] = 2
print("Berhasil load")
```

```python
# Grafik perbandingan durasi seluruh data dikelompokkan berdasarkan jenis teks
# Audio "zero"
plt.plot(durasi[0:11], label="Audio 0")
# Audio "one"
plt.plot(durasi[11:22], label="Audio 1")
# Audio "two"
plt.plot(durasi[22:34], label="Audio 2")
plt.xlabel("Track number")
plt.ylabel("Durasi")
plt.legend()
plt.savefig("data/grafik-durasi-kelompok.png")
```

```python
# Grafik perbandingan durasi seluruh data set
plt.plot(durasi)
plt.xlabel("Track number")
plt.ylabel("Durasi")
plt.savefig("data/grafik-durasi.png")
```

```python
# Membuat unit uji
# Classifier : bayes
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(cost_matrix, y)
```

```python
# Fungsi prediksi teks audio berdasarkan data training
def predict_result(file_test):
    audio_test, scl1 = librosa.load(file_test)
    mfcc = librosa.feature.mfcc(audio_test, scl1)
    distanceTest = []
    for i in range(len(files)):
        audio_comp, scl_comp = librosa.load(train_path + "/" + files[i])
        mfcc1 = librosa.feature.mfcc(audio_comp, scl_comp)
        _,_,cost = dtw(mfcc.T, mfcc1.T, dist=lambda x, y: np.linalg.norm(x-y, ord=1))

        distanceTest.append(cost)
    hasil = gnb.predict([distanceTest])
    return hasil, distanceTest
```

```python
# Visualisasi perbandingan hasil cost dari dtw terhadap tiap audio number
def visualize(distance, audio_number):
    # Audio "zero"
    plt.plot(distance[0:11], label="Audio 0")
    # Audio "one"
    plt.plot(distance[11:22], label="Audio 1")
    # Audio "two"
    plt.plot(distance[22:34], label="Audio 2")
    plt.xlabel("Track number")
    plt.ylabel("Distance")
    plt.legend()
    plt.savefig("data/distance-test-"+str(audio_number)+".png")
```
